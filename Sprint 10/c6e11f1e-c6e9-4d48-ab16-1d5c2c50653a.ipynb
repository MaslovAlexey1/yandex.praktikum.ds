–í —ç—Ç–æ–º –∑–∞–¥–∞–Ω–∏–∏ –≤—ã –º–æ–∂–µ—Ç–µ –∑–∞–ø–∏—Å—ã–≤–∞—Ç—å —Ñ–æ—Ä–º—É–ª—ã –≤ *Jupyter Notebook.*

–î–ª—è –ø—Ä–∏–º–µ—Ä–∞ –º—ã –∑–∞–ø–∏—Å–∞–ª–∏ —Ñ–æ—Ä–º—É–ª—ã –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –ú–æ–∂–µ—Ç–µ –∏—Ö —Å–∫–æ–ø–∏—Ä–æ–≤–∞—Ç—å –∏ –æ—Ç—Ä–µ–¥–∞–∫—Ç–∏—Ä–æ–≤–∞—Ç—å, —á—Ç–æ–±—ã —Ä–µ—à–∏—Ç—å –∑–∞–¥–∞—á—É.

–†–∞–±–æ—Ç–∞—Ç—å –≤ *LaTeX* –Ω–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ.

–û–±–æ–∑–Ω–∞—á–µ–Ω–∏—è:

- $X$ ‚Äî –º–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–Ω—É–ª–µ–≤–æ–π —Å—Ç–æ–ª–±–µ—Ü —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –µ–¥–∏–Ω–∏—Ü)

- $y$ ‚Äî –≤–µ–∫—Ç–æ—Ä —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞

- $P$ ‚Äî –º–∞—Ç—Ä–∏—Ü–∞, –Ω–∞ –∫–æ—Ç–æ—Ä—É—é —É–º–Ω–æ–∂–∞—é—Ç—Å—è –ø—Ä–∏–∑–Ω–∞–∫–∏

- $w$ ‚Äî –≤–µ–∫—Ç–æ—Ä –≤–µ—Å–æ–≤ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ (–Ω—É–ª–µ–≤–æ–π —ç–ª–µ–º–µ–Ω—Ç —Ä–∞–≤–µ–Ω —Å–¥–≤–∏–≥—É)

–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è:

$$
a = Xw
$$

–ó–∞–¥–∞—á–∞ –æ–±—É—á–µ–Ω–∏—è:

$$
w = \arg\min_w MSE(Xw, y)
$$

–§–æ—Ä–º—É–ª–∞ –æ–±—É—á–µ–Ω–∏—è:

$$
w = (X^T X)^{-1} X^T y
$$

**–û—Ç–≤–µ—Ç:** –Ω–µ –∏–∑–º–µ–Ω–∏—Ç—Å—è

**–û–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ:**
–ü–æ—Å–º–æ—Ç—Ä–∏–º –∫ —á–µ–º—É –º—ã –ø—Ä–∏–¥–µ–º —É–º–Ω–æ–∂–∏–≤ X –Ω–∞ –æ–±—Ä–∞—Ç–∏–º—É—é –º–∞—Ç—Ä–∏—Ü—É A<br>
–£—á–∏—Ç—ã–≤–∞—è, —á—Ç–æ $ (ABC)^{-1} = C^{-1}B^{-1}A^{-1} $ –∏ $ (AB)^T = B^T A^T $ <br><br>
$$
w_A = ((X A)^T (X A))^{-1} (X A)^T y = (A^T X^T X A)^{-1} A^T X^T y = A^{-1} X^{-1} (X^T)^{-1} (A^T)^{-1} A^T X^T y = ...
$$<br>
–¢–∞–∫ –∫–∞–∫ –º–∞—Ç—Ä–∏—Ü–∞ A –æ–±—Ä–∞—Ç–∏–º–∞—è, —Ç–æ $ (A^T)^{-1} A^T = E $ –∏:<br><br>
$$
... = A^{-1} X^{-1} (X^T)^{-1} E  X^T y = A^{-1} (X^T X)^{-1} X^T y = A^{-1} w
$$

## 4. –ü—Ä–æ–≤–µ—Ä–∫–∞ –∞–ª–≥–æ—Ä–∏—Ç–º–∞
